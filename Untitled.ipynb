```python
def get_eye_diameters(file_path):
    # df = pd.read_csv("000/exports/000/pupil_positions.csv")
    df = pd.read_csv(file_path)

    df_max_confidence = df.loc[df.groupby(['world_index', 'eye_id'])['confidence'].idxmax()]

    df_max_confidence = df_max_confidence[['world_index', 'eye_id', 'diameter']]
    pivot_df = df_max_confidence.pivot(index='world_index', columns='eye_id', values='diameter')

    return pivot_df

def plot_eye_diameters(pivot_df):
    # Calculate the mean of eye diameters
    pivot_df['mean_diameter'] = pivot_df.mean(axis=1)

    # Convert frame index to seconds
    pivot_df['time'] = pivot_df.index / FPS

    # Plotting
    plt.figure(figsize=(15, 6))  # Increased figure width
    #plt.plot(pivot_df['time'], pivot_df[0], label='Eye 0')
    #plt.plot(pivot_df['time'], pivot_df[1], label='Eye 1')
    plt.plot(pivot_df['time'], pivot_df['mean_diameter'], label='Mean Diameter', linestyle='-')

    plt.title('Eye Diameters Over Time')
    plt.xlabel('Time (seconds)')
    plt.ylabel('Diameter')
    plt.legend()
    plt.show()
pupil_dilations = get_eye_diameters(pupil_positions_path)
plot_eye_diameters(pupil_dilations)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
import ipywidgets as widgets

# Function to load gaze data
def get_gaze_positions(file_path):
    df = pd.read_csv(file_path)
    df_max_confidence = df.loc[df.groupby('world_index')['confidence'].idxmax()]
    return df_max_confidence[['norm_pos_x', 'norm_pos_y']]
def plot_heatmap_of_gaze(index1,index2):
    # Convert the normalized positions to pixel coordinates (assuming a certain image size)
    current_data = gaze_positions.iloc[index1:index2]
    
    image_size = [1920, 1080]  # Example image size (width, height)


    # Perform Kernel Density Estimation
    x = current_data['norm_pos_x'] * image_size[0]
    y = current_data['norm_pos_y'] * image_size[1]
    kde = gaussian_kde([x, y], bw_method='silverman')

    # Create a grid of points for evaluation
    x_grid, y_grid = np.meshgrid(np.linspace(0, image_size[0], 100), np.linspace(0, image_size[1], 100))
    grid_points = np.vstack([x_grid.ravel(), y_grid.ravel()])

    # Evaluate the KDE on the grid
    z = kde(grid_points).reshape(x_grid.shape)

    # Plot the heatmap
    plt.figure(figsize=(12, 6))
    plt.imshow(z, origin='lower', aspect='auto', extent=[0, image_size[0], 0, image_size[1]], cmap='hot')
    plt.colorbar(label='Density')
    plt.xlabel('X coordinate')
    plt.ylabel('Y coordinate')
    plt.title('Gaze Position Heatmap')
    # plt.show()
# Load the gaze positions
gaze_positions = get_gaze_positions(gaze_positions_path)

slider1 = widgets.IntSlider(
    value=0,  # Initial value
    min=1,  # Minimum value
    max=len(gaze_positions),  # Maximum value
    step=1,  # Step size
    description='Index:',  # Description label
    continuous_update=False  # Update only when the slider is released
)
slider2 = widgets.IntSlider(
    value=len(gaze_positions),  # Initial value
    min=1,  # Minimum value
    max=len(gaze_positions),  # Maximum value
    step=1,  # Step size
    description='Index:',  # Description label
    continuous_update=False  # Update only when the slider is released
)
widgets.interactive(plot_heatmap_of_gaze, index1=slider1, index2=slider2)

```

```python
from config import IMG_SIZE
from detect_aruco_video import get_data_from_csvs, get_iters_from_lists
import cv2

from utils import estimate_gaze_pos
def get_marker_count_per_frame(marker_positions):
    # Assuming marker_positions is a list where each element is a list of markers for that frame
    print(marker_positions)
    return [len(markers) for markers in marker_positions]

def get_frames(video_path):
    
    # Open the video file
    cap = cv2.VideoCapture(video_path)
    frames = []
    number_of_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)

    # Check if video opened successfully
    if not cap.isOpened():
        print("Error opening video file")
    while cap.isOpened():
        # Capture frame-by-frame
        ret, frame = cap.read()
        if ret:
            # Append the frame to the list
            frames.append(frame)
        else:
            break


    # When everything done, release the video capture object
    cap.release()
    return frames, number_of_frames, frame_width, frame_height, fps

def get_heatmapped_gaze_positions(bins = 40):
    heatmap_gaze_positions = []
    for g in gaze_positions:
        temp = estimate_gaze_pos(g)
        normalized_gaze_x = int((g[0] / IMG_SIZE[0]) * (bins - 1))
        normalized_gaze_y = int((g[1]/ IMG_SIZE[1]) * (bins - 1))
        heatmap_gaze_positions.append((normalized_gaze_x,normalized_gaze_y))
    return heatmap_gaze_positions

def get_continous_blinks_list(blinks):

    blink_values = np.zeros(number_of_frames)
    for blink in blinks:
        start, end = blink
        blink_values[start:end] = 1
    return blink_values

blinks, eye_diameters, gaze_positions, head_poses, marker_positions = get_data_from_csvs(blinks_path,
                                                                                            gaze_positions_path,
                                                                                            headpose_tracker_poses_path,
                                                                                            marker_detections_path,
                                                                                            pupil_positions_path,
                                                                                            world_timestamps_path)

# heatmap_gaze_positions = get_heatmapped_gaze_positions()

marker_counts_per_frame = get_marker_count_per_frame(marker_positions)
frames, number_of_frames, frame_width, frame_heigth, fps = get_frames(video_path)
continous_blinks_list = get_continous_blinks_list(blinks)
```

```python
import cv2
WIDTH = 480
HEIGHT = 320
VIDEO_WIDTH, VIDEO_HEIGHT = 1280, 720
IMG_SIZE = (1200, 1200)
FONT = cv2.FONT_HERSHEY_SIMPLEX
FONT_SCALE = 0.5
TEXT_COLOR = (0, 0, 0)  # Black color
TEXT_THICKNESS = 1
TEXT_LINE_SPACING = 20
FOV_THICKNESS = 2
FOV_COLOR = (1,0,0)
MARKER_THICKNESS=2
MARKER_VISIBLE_COLOR = (0,0,1)
MARKER_NOT_VISIBLE_COLOR = (0,1,1)
GAZE_CIRCLE_COLOR = (0,1,0)
GAZE_CIRCLE_RADIUS = 15
MARKER_ENLARGEMENT_RATE = 2
EYE_DIAMETER_CIRCLE_THICKNESS = 2
EYE_DIAMETER_NORMALIZING_FACTOR = 50
EYE_DILATION_COLOR = (1,0,0)
BLINK_LIST_LENGTH = 3
NUMBER_OF_PAST_POSES_TO_VISUALIZE = 10  # number of past poses to store and visualize
GAZE_KERNEL_FOR_HEATMAP = 5
FPS = 24
```

```python
from utils import estimate_gaze_pos


def get_translated_view_rectangle(pitch, roll, yaw):
    center_y = int((90 - pitch) / 180 * IMG_SIZE[1])
    center_x = int((90 + yaw) / 180 * IMG_SIZE[0])
    center = (center_x, center_y)
    rect = np.array([
        [-WIDTH / 2, -HEIGHT / 2],
        [WIDTH / 2, -HEIGHT / 2],
        [WIDTH / 2, HEIGHT / 2],
        [-WIDTH / 2, HEIGHT / 2]
    ])
    
    rotation_matrix = cv2.getRotationMatrix2D((0, 0), roll, 1)
    rotated_rect = np.dot(rect, rotation_matrix[:, :2].T)
    translated_rect = rotated_rect + center
    return center, translated_rect
def visualize_with_matplotlib(ax, translated_rect, markers, gaze_position, eye_diameter):
    # Create a blank image
    # img = np.ones((IMG_SIZE[1], IMG_SIZE[0], 3))
    
    # # Double the figure size
    # ax.imshow(img)

    for j in range(4):
        start_point = translated_rect[j]
        end_point = translated_rect[(j + 1) % 4]
        # Adjusted line width
        ax.plot([start_point[0], end_point[0]], [start_point[1], end_point[1]], color=FOV_COLOR, linewidth=FOV_THICKNESS * 2)

    # Draw Markers
    for marker_id, marker_center, marker_corners in markers:
        for j in range(4):
            start_point = marker_corners[j]
            end_point = marker_corners[(j + 1) % 4]
            # Adjusted line width
            ax.plot([start_point[0], end_point[0]], [start_point[1], end_point[1]], color=MARKER_VISIBLE_COLOR, linewidth=MARKER_THICKNESS * 2)

    # Draw Gaze and Dilation Circle
    eye_dilation_radius = int(eye_diameter / EYE_DIAMETER_NORMALIZING_FACTOR * GAZE_CIRCLE_RADIUS)
    # Adjusted circle radius
    circle1 = plt.Circle((gaze_position[0], gaze_position[1]), GAZE_CIRCLE_RADIUS * 2, color=GAZE_CIRCLE_COLOR, fill=False)
    circle2 = plt.Circle((gaze_position[0], gaze_position[1]), eye_dilation_radius * 2, color=EYE_DILATION_COLOR, fill=False)
    ax.add_artist(circle1)
    ax.add_artist(circle2)

    # Set limits and aspect ratio
    ax.set_xlim(0, IMG_SIZE[0])
    ax.set_ylim(IMG_SIZE[1], 0)
    # ax.set_aspect('equal')

    # plt.show()

def plot_eye_diameters(ax, eye_diameters, index, range_around_index=100):

    # Calculate start and end indices
    start_index = max(0, index - range_around_index)
    end_index = min(len(eye_diameters), index + range_around_index + 1)

    # Extracting left and right eye diameters
    left_eye_diameters = [diameter[0] for diameter in eye_diameters[start_index:end_index]]
    right_eye_diameters = [diameter[1] for diameter in eye_diameters[start_index:end_index]]

    # Calculate mean diameters
    mean_diameters = [(left + right) / 2 for left, right in zip(left_eye_diameters, right_eye_diameters)]

    # Generate x-axis values (frame numbers)
    frame_numbers = list(range(start_index, end_index))

    # Plotting
    # plt.plot(frame_numbers, left_eye_diameters, marker='o', label='Left Eye')
    # plt.plot(frame_numbers, right_eye_diameters, marker='o', label='Right Eye')
    ax.plot(frame_numbers, mean_diameters, marker='x', linestyle='-', label='Mean Diameter')
    ax.axvline(x=index, color='red', linestyle='--')  # Mark the specified index
    ax.set_title(f'Eye Diameters around Frame {index}')
    ax.set_xlabel('Frame Number')
    ax.set_ylabel('Eye Diameter')
    ax.legend()
    ax.grid(True)
    # ax.show()

def plot_markers_over_time(ax, marker_counts, index, range_around_index=100):
    start_index = max(0, index - range_around_index)
    end_index = min(len(marker_counts), index + range_around_index + 1)
    
    frame_numbers = list(range(start_index, end_index))
    marker_values = marker_counts[start_index:end_index]
    
    ax.plot(frame_numbers, marker_values, linestyle='-', label='Markers over time')
    ax.axvline(x=index, color='red', linestyle='--')
    ax.set_title(f'Markers Count around Frame {index}')
    ax.set_xlabel('Frame Number')
    ax.set_ylabel('Number of Markers')
    ax.set_ylim(-0.5, 8.5)  # Adjust y-axis limits

    ax.legend()
    ax.grid(True)
    # ax.show()

def plot_blinks(ax, continous_blinks, index, range_around_index=300):
    # Calculate the frame range to display
    start_index = max(0, index - range_around_index)
    end_index = min(len(eye_diameters), index + range_around_index + 1)

    # Create a list for x-axis (frame number) and y-axis (blink presence)
    frame_numbers = list(range(start_index, end_index))
    continous_blinks = continous_blinks[start_index:end_index]
    ax.plot(frame_numbers, continous_blinks, linestyle='-', label='Blinks over time')
    ax.axvline(x=index, color='red', linestyle='--')  # Mark the specified index
    ax.set_title(f'Blinks around frame {index}')
    ax.set_xlabel('Frame Number')
    ax.set_ylabel('Eye Diameter')
    ax.set_ylim(-0.5, 1.5)  # Corrected y-axis limits setting

    ax.legend()
    ax.grid(True)
    # ax.show()
    # Mark the frames where blinks occur

def get_heatmap_gaze_positions(gaze_positions, head_poses, frame_index, range_of_frames):
    sub_head_poses = head_poses[frame_index-range_of_frames:frame_index]
    sub_gaze_positions = gaze_positions[frame_index-range_of_frames:frame_index]
    heatmap_gaze_positions = []
    for i in range(len(sub_head_poses)):
        pitch, yaw, roll = sub_head_poses[i]
        fov_center, fov_rectangle = get_translated_view_rectangle(pitch, roll, yaw)

        heatmap_gaze_positions.append(estimate_gaze_pos(fov_center,sub_gaze_positions[i],roll))
    return heatmap_gaze_positions
def plot_gaze_heatmap(ax, head_poses, gaze_positions, frame_index, range_of_frames):
    # Assuming the input head_poses and gaze_positions are available as global variables

    # Get the heatmap gaze positions using your function
    heatmap_gaze_positions = get_heatmap_gaze_positions(gaze_positions, head_poses, frame_index, range_of_frames )

    # Convert the heatmap gaze positions to x and y arrays
    if len(heatmap_gaze_positions):
        x, y = zip(*heatmap_gaze_positions)
    else:
        x,y = 0,0

    # Perform Kernel Density Estimation
    kde = gaussian_kde([x, y], bw_method='silverman')

    # Create a grid of points for evaluation
    x_grid, y_grid = np.meshgrid(np.linspace(min(x), max(x), 100), np.linspace(min(y), max(y), 100))
    grid_points = np.vstack([x_grid.ravel(), y_grid.ravel()])

    # Evaluate the KDE on the grid
    z = kde(grid_points).reshape(x_grid.shape)

    # Plot the heatmap
    # ax.figure(figsize=(12, 6))
    ax.imshow(z, origin='lower', aspect='auto', extent=[min(x), max(x), min(y), max(y)], cmap='hot')
    # ax.colorbar(label='Density')
    ax.set_xlabel('X coordinate')
    ax.set_ylabel('Y coordinate')
    ax.set_title('Gaze Position Heatmap')
    # ax.show()

```
